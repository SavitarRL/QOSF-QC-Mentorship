### Quantum Solution

Answer

```python
from solution import QuantumSolution
quantum_sol = QuantumSolution(k = 11, list_n = [3, 10, 14, 7, 12, 20, 24 ])
ans = quantum_sol.less_than_k()
print(ans[0])
```

```python
type(ans[1])
```

Plotting histogram of probabilities

```python
quantum_sol.plot_answer_histogram(ans[1])
```

Circuit for state preparation

```python
quantum_sol.init_qc.decompose(reps = 7).draw(output = "mpl")
```

Circuit of the grover oracle

```python
quantum_sol.grover_operator.decompose().draw(output = "mpl")
```

Resources used

```python
quantum_sol.get_resources()
```

Exhaustive solutions

```python
for k in range(1,26):
    quantum_sol = QuantumSolution(k=k, list_n=[3, 10, 14, 7, 12, 20, 24])
    ans = quantum_sol.less_than_k()
    print("k = {} : {}".format(k, ans[0]))
```
